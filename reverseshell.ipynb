## HTTP Callback Test (DataSphere Compatible)
This will send requests to your ngrok server.

```python
import requests, time

URL = "https://2daa-2409-40e3-400e-3c03-ad52-2bd5-4857-5dd7.ngrok-free.app/"

while True:
    try:
        r = requests.get(URL)
        print("Status:", r.status_code)
        print("Response:", r.text[:100])
    except Exception as e:
        print("Error:", e)

    time.sleep(5)
```
